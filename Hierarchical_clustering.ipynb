```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.datasets import make_blobs

# Generate sample data (change as needed)
X, _ = make_blobs(n_samples=50, centers=3, random_state=42)

# Perform hierarchical clustering using 'ward' method
linked = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Find the largest gap in the linkage matrix to determine the optimal number of clusters
distances = linked[:, 2]  # Extract the distances between clusters
gaps = np.diff(distances)  # Calculate gaps between successive distances

# Find the largest gap, which indicates where to cut the dendrogram
largest_gap_index = np.argmax(gaps)  # Index of the largest gap
optimal_distance = distances[largest_gap_index + 1]  # The distance to cut the dendrogram

# Get cluster labels by cutting the dendrogram at the optimal distance
labels = fcluster(linked, t=optimal_distance, criterion='distance')

# Number of clusters found
n_clusters = len(np.unique(labels))
print(f'Optimal number of clusters found: {n_clusters}')

# Plotting the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title(f'Clusters found using Hierarchical Clustering (n_clusters={n_clusters})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

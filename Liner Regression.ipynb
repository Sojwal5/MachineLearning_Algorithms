```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv("homeprices.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df.area,df.price,color='red',marker='+')
plt.xlabel("Ares(in sq)")
plt.ylabel("price(US)")
```

```python
X=df[["area"]] #X WILL BE IN 2D AS FEATURE ARE MULTIPE SO 1 D IN FEATURES AND OTHER 2D FOR VALUES
y=df['price']
model=LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25) 
```

```python
X_train
```

```python
y_train #first train wale se line banayi  then test wale se accuracy dekhi
```

```python
model.fit(X_train,y_train)  # pahale ytest  predict kiya then ytest jo actual tha usse minus kiya
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
# Plotting both training and test points
plt.scatter(X_train, y_train, color='green', label='Training Data')  # Training points in green
plt.scatter(X_test, y_test, color='blue', label='Test Data')  # Test points in blue

# Plotting the regression line (predictions on the test data)
plt.plot(X_test, y_pred, color='red', label='Prediction Line')

# Labels and title
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression")

# Display legend
plt.legend()

# Show the plot
plt.show()
```

```python
col_useful=['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df= df[col_useful]

#selecting colums that we want to use and dropping the another columns
```

```python
selected_columns = ['Rooms', 'Distance','Bathroom','Bedroom2' ,'Car', 'Landsize','BuildingArea','Price']
correlation_matrix = df[selected_columns].corr()
print(correlation_matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Selected Features')
plt.show()
#as the corelation between the the features is very low close to 0 not near -1 and 1 so linear regression is not the option 
#polynomial regesiion might help as the releation will be a random curve
#we can also apply overfitting techniques
```

```python
#filling null values
#1.Replacing values with 0
#2.Replacing values with mean
#3.dropping the complete row

# Some feature's missing values can be treated as zero (another class for NA values or absence of that feature)
# like 0 for Propertycount, Bedroom2 will refer to other class of NA values
# like 0 for Car feature will mean that there's no car parking feature with house
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)

# other continuous features can be imputed with mean for faster results since our focus is on Reducing overfitting
# using Lasso and Ridge Regression
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
#One hot encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
dataset = pd.DataFrame(df)

# Replace True/False with 1/0 for all columns
dataset = dataset.replace({True: 1, False: 0})
```

```python
dataset.head()
```

```python
dataset.tail()
dataset.dtypes
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
lin_model = LinearRegression()
```

```python
lin_model.fit(train_X,train_y)
```

```python
lin_model.score(test_X, test_y)
```

```python
lin_model.score(train_X, train_y)
#It clearly shows that it is overfitting 
#Good accuracy in train but bad in test 
```

```python
predict_y=lin_model.predict(test_X)
mse=mean_squared_error(test_y,predict_y)
r2=r2_score(test_y, predict_y)
print(mse)
print(r2)
#r2 score is very less
```

```python
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
predict_y1=lasso_reg.predict(test_X)
```

```python
predict_y=lin_model.predict(test_X)
mse=mean_squared_error(test_y,predict_y1)
r2=r2_score(test_y, predict_y1)
print(mse)
print(r2)
#r2 score is increased by lasso regression
```

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```

```python
predict_y2=ridge_reg.predict(test_X)
mse=mean_squared_error(test_y,predict_y2)
r2=r2_score(test_y, predict_y2)
print(mse)
print(r2)
#therefor by using ridge and lasso we have increased the performance of the model
```
